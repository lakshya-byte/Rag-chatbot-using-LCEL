### message history

```python
from langchain_community.chat_message_histories import ChatMessageHistory
from langchain_core.chat_history import BaseChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory

store = {}


def get_session_history(session_id: str) -> BaseChatMessageHistory:
    if session_id not in store:
        store[session_id] = ChatMessageHistory()
    return store[session_id]

with_message_history = RunnableWithMessageHistory(model , get_session_history)
```

```python
config = {"configurable":{"session_id":"chat1"}}
```

```python
response = with_message_history.invoke(
    [HumanMessage(content="Hi, my name is lakshya and i am web developer")], 
    config=config
)
response
```

```python
config1 = {"configurable":{"session_id":"chat1"}}
response = with_message_history.invoke(
    [HumanMessage(content="what is my name")], 
    config=config1
)
print(response.content)
```

### managing the conversation history 

One important concept to understand when buidling chatbots is how to mnage conversation history. If left unmanaged , the list of messages will grow unbounded and potentially overflow the context window of the LLM. Therefore, it is important to add a step that limits the size of messages you are passing in.

```python
pip install tiktoken
```

```python
# If you don't have tiktoken installed, run this in a separate cell first:
# %pip install tiktoken

import tiktoken
from langchain_groq import ChatGroq
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage, trim_messages

# 1. Initialize your Groq model
# (Make sure your groq_api_key variable is defined above this)
model = ChatGroq(api_key=groq_api_key, model="llama-3.3-70b-versatile")

# 2. THE FIX: Define a custom token counter
def fast_proxy_counter(messages: list) -> int:
    encoding = tiktoken.get_encoding("cl100k_base")
    num_tokens = 0
    for msg in messages:
        # Handle different message formats safely
        content = msg.content if isinstance(msg.content, str) else str(msg.content)
        num_tokens += len(encoding.encode(content))
    return num_tokens

# 3. Wire the custom counter into the trimmer
trimmer = trim_messages(
    max_tokens=70,
    strategy="last",
    token_counter=fast_proxy_counter,  # <--- We pass our function here, NOT 'model'
    include_system=True,
    allow_partial=False,
    start_on="human"
)

# 4. Your test messages
messages = [
    SystemMessage(content="you are a good assistant"),
    HumanMessage(content="Hi i am bob"),
    AIMessage(content="Hi Bob! How can I help you today?"),
    HumanMessage(content="I am a web developer"),
    AIMessage(content="That's great! What technologies do you work with?")
]

# 5. Execute
try:
    trimmed_history = trimmer.invoke(messages)
    print("✅ SUCCESS! Trimmed History:\n")
    for m in trimmed_history:
        print(f"[{m.type.upper()}]: {m.content}")
except Exception as e:
    print(f"❌ ERROR: {e}")
```

```python
from operator import itemgetter
from langchain_core.runnables import RunnablePassthrough

chain=(
   RunnablePassthrough.assign(
    messages = itemgetter("messages")  | trimmer
   ) | prompt | model
)

response = chain.invoke(
 {   "messages": messages + [HumanMessage(content="what icecream do i like")],
    "language": "english"}
)

print(response)
```

## Lets wrap this thing in the message history

```python
with_message_history = RunnableWithMessageHistory(
    chain,
    get_session_history,
    input_messages_key="messages",
)
config={"configurable":{"session_id":"chat5"}}
```

```python
response = with_message_history.invoke(
    {
        "messages": [
            HumanMessage(content="what math problem do i ask")
        ],
        "language": "english"
    },
    config = config
)
response.content
```
